```python
import polars as pl
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.utils.class_weight import compute_class_weight

import torch
from torch.nn import CrossEntropyLoss

from datasets import Dataset
from transformers import AutoTokenizer, AutoModelForSequenceClassification, TrainingArguments, Trainer
```

```python
DATA_PATH = "./data/cicids2017_after_EDA.parquet"
LABEL_COL = "Label"
TEXT_COL = "Text"
```

```python
df = pl.read_parquet(DATA_PATH)
df.shape
```

```python
# serialize all feature columns except label
feature_cols = [c for c in df.columns if (c != LABEL_COL and c != "Timestamp")]
df = df.with_columns(
    pl.concat_str(
        [pl.col(c).cast(pl.Utf8).alias(c) for c in feature_cols],
        separator=" "
    ).alias(TEXT_COL)
)
```
