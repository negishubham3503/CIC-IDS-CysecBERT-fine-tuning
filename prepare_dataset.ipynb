```python
import pandas as pd, glob, chardet
```

```python
# Load all CSV files
files = glob.glob("data/*.csv")  # or specify paths: ["file1.csv", "file2.csv"]
dataframes = {f: pd.read_csv(f, nrows=0) for f in files}  # nrows=0 = headers only

# Compare columns against the first file
reference = list(dataframes[files[0]].columns)
print(f"Reference columns ({files[0]}): {reference}\n")

all_match = True
for file, df in dataframes.items():
    cols = list(df.columns)
    if cols != reference:
        all_match = False
        print(f"❌ MISMATCH — {file}")
        print(f"   Missing : {set(reference) - set(cols)}")
        print(f"   Extra   : {set(cols) - set(reference)}")
        print(f"   Order diff: {cols != reference and set(cols) == set(reference)}")
    else:
        print(f"✅ Match   — {file}")

print("\n✅ All columns match!" if all_match else "\n❌ Some files differ.")
```

```python
def read_csv_auto(f):
    with open(f, "rb") as raw:
        encoding = chardet.detect(raw.read())["encoding"]
    print(f"📄 {f} — detected encoding: {encoding}")
    return pd.read_csv(f, encoding="cp1252", low_memory=False)

dfs = [read_csv_auto(f) for f in files]
cic_ids_df = pd.concat(dfs, ignore_index=True)

cic_ids_df.to_csv("data/cic_ids_df_combined.csv", index=False)
print(f"✅ Merged {len(files)} files → {len(cic_ids_df)} total rows saved to cic_ids_df_combined.csv")
```
