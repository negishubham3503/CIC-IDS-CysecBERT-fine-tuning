```python
import polars as pl
```

```python
%load_ext memory_profiler
```

```python
cic_df = pl.read_csv("data/cic_ids_df_combined.csv")
cic_df.describe()
```

```python
cic_df.head()
```

```python
def clean_column_names(cols):
    # Remove leading/trailing spaces, replace '.' and ' ' with '_'
    return [c.strip().replace('.', '').replace(' ', '_') for c in cols]

# Clean column names
new_names = clean_column_names(list(cic_df.columns))
cic_df.columns = new_names

# Remove duplicate header lines
if "Flow_ID" in cic_df.columns:
    cic_df = cic_df.filter(cic_df["Flow_ID"] != "Flow_ID")

# Remove full duplicates
%memit
%time
before = cic_df.shape
cic_df = cic_df.unique()
after = cic_df.shape
print(f"Removed {before[0]-after[0]} duplicate rows")

# Quick peek
cic_df.head(3)
```

```python
drop_cols = [c for c in cic_df.columns if any(x in c.lower() for x in ['flow_id', 'source_ip', 'destination_ip'])]
cic_df = cic_df.drop(drop_cols)
```

```python
# Find duplicates by all fields except Flow_ID
dupes_mask = cic_df.is_duplicated()

num_soft_dupes = dupes_mask.sum()
print(f"Rows with identical feature values but different Flow_ID: {num_soft_dupes}")

# Optionally drop them (keep first occurrence only)
df_nodupes = cic_df.unique()

print(f"Shape after dropping these soft duplicates: {df_nodupes.shape}")
```

```python
df_nodupes.head()
```

```python
# Select all columns except Flow_ID for duplicate finding
cols_for_duplicate = [c for c in df_nodupes.columns if c != "Timestamp"]

# Find duplicates by all fields except timestamp
dupes_mask = df_nodupes.select(cols_for_duplicate).is_duplicated()

num_soft_dupes = dupes_mask.sum()
print(f"Rows with identical feature values but different timestamp: {num_soft_dupes}")

# Optionally drop them (keep first occurrence only)
df_final = df_nodupes.unique(subset=cols_for_duplicate)

print(f"Shape after dropping these soft duplicates: {df_final.shape}")
```

```python
len(df_final.select("Source_Port").unique())
```

```python
# Drop ID/timestamp columns from grouping
exclude_cols = {"Timestamp", "Label", "Source_Port", "Destination_Port"}
df = df_final
group_cols = [c for c in df.columns if c not in exclude_cols and not ("ip" in c.lower())]

# Group, count unique labels per feature-row
grouped = (
    df.group_by(group_cols)
      .agg([
          pl.col("Label").n_unique().alias("n_labels"),
          pl.col("Label").unique().alias("labels_found"),
          pl.len().alias("n_rows")
      ])
      .filter(pl.col("n_labels") > 1)  # only conflicting label sets
)

print(f"Number of unique feature combinations with conflicting labels: {grouped.shape[0]}")
grouped.head(10)  # Show a sample
```

```python
# Save Polars DataFrame as Parquet (all columns, compressed, fast)
df.write_parquet("cicids2017_preprocessed.parquet")
print("DataFrame saved to cicids2017_preprocessed.parquet")
```
